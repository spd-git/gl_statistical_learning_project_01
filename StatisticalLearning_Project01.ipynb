### 2. Print all the teams which have started playing between 1930-1980. 
(5 points) [Jump to Result](#result2)

```python
#With this we know that there are no values other than strings in the column
df.Debut.apply(lambda val: type(val) == str).isnull().values.any()
```

```python
#This function converts data like '1999-00' to a np.array [1999, 2000].
def debut_cleanup(val):
    val = val.split('-')
    if (len(val) == 2):
        if val[0][2:] == "99":
            val[1]=str(int(val[0][:2])+1)+val[1]
        else:
            val[1]=val[0][:2]+val[1]
    return np.array(list(map(int, val)))
```

```python
df['Debut'] = df.Debut.apply(debut_cleanup)
```

<a id='result2'>Result 2:</a>

```python
# Incase a team debuts in 1929-30 or 1980-81, the record will be shown
df[df.apply(lambda x:True if (x.Debut.max()>=1930 and x.Debut.min()<=1980) else False , axis=1)][['Team']]
```

### 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. 
(5 points) [Jump to Result](#result4)

<a id='result4'>Result 4:</a>

```python
def goal_diff_count(row):
    return row.GoalsFor - row.GoalsAgainst
```

```python
df['GoalsDiff'] = df.apply(goal_diff_count,axis=1)
```

```python
# Method 1
df.loc[df.GoalsDiff.isin(df.GoalsDiff.quantile([0,1]))][['Team','GoalsDiff']]
```

```python
# Method 2 with set operations
df[['Team', 'GoalsDiff']].loc[df['GoalsDiff'].isin({df['GoalsDiff'].max(),
                             df['GoalsDiff'].min()})]
```

### 5. Create a new column with name “Winning Percent” and append it to the data set 
(5 points) [Jump to Result](#result5)

Percentage of Winning = (GamesWon / GamesPlayed)*100
If there are any numerical error, replace it with 0%
Print the top 5 teams which has the highest Winning percentage

<a id='result5'>Result 5:</a>

```python
def winning_percentage(row):
    return (row.GamesWon / row.GamesPlayed)*100 if row.GamesPlayed != 0 else 0
```

```python
df['Winning Percent'] = df.apply(winning_percentage,axis=1)
```

```python
df.sort_values(by=['Winning Percent'],ascending=False).iloc[:5]
```
